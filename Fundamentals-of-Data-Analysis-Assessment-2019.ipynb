# 2. Regression

In this section we will discuss regression. Regression is one of the main areas of research in Data Analysis. We will look at plotting the **total_bill** versus **tip**. 

Once this is plotted we will look at finding the best fit linear regression line and plotting this line on a scatter plot.

References:<br>
http://www.stat.yale.edu/Courses/1997-98/101/linreg.htm<br>
https://seaborn.pydata.org/tutorial/regression.html<br>


```python
# The relationship between total_bill and tips is plotted

plt.scatter(df.loc[:,'total_bill'], df.loc[:,'tip'])
plt.title('Tip vs Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
```

The linear regression is a line that best fits the two plotted variables, the Y variable is the dependant variable i.e the value of *tip* is dependent on *total_bill*. 

The best line is determined by the following summation formula this method is called the **least squares fitting** formula. This is the cost of the line

$\sum_i  (y_i - mx_i - c)^2$

$y_i - mx_i - c$ is determined mathematically from the formula $y_i = mx_i + c$ which is a straight line with a slope of m, the c constant value is the value of y when x = 0

- i = number of x & y values that have to be squared and summed or the number of dots on the scatter plot above<br>
- x = total_bill that y (tip) is dependent on<br>
- y = tip<br>
- m = slope of the line<br>
- c = constant value line hits the Y axis when x = 0<br>

The formula is the sum of all the distances each value of x & y (i) is away from the line, the (i) or point on the plot can be either above or below the line, we want to find a best fit line. For each value if i this summation is performed. The best line will have the minimum distance from the points on the plot as a whole. Because some points will be above the line and some points will be below the line, there would be negative numbers this will not yield the desired result. Also rather that taking the absolute negative value and making it positive, it is easier to simply squared all the numbers . If you square a negative or even positive number its a positive number. 

For a particular line you need to find out the distance each dot is from the line and sum this total, the best line will be the lowest number of the sum total of each dot's distance from the line. This is the lowest cost (m, c). For example if all the dots were on a straight line then the cost would be zero.
