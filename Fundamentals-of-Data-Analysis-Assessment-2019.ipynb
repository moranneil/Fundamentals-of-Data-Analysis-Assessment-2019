Regression can be used to to see if there is a relationship between an independent variable and dependent variable, the regression statistics can be used to predict the value of the dependent variable when the independent variable is known. We will look at plotting the **total_bill** versus **tip**. 

In regression analysis the independent variable is known as the x-value and the dependent variable is known as the y-value

From the tips dataset the variables assigned to x & y are given below

x = total_bill<br>
y = tip<br>

x is the independent variable and y is the dependent variable.

```python
#The relationship between total_bill and tips is plotted with a scatter plot
#This is a magic command to ensure the plot is displayed in Jupyter notebook
%matplotlib inline
#The plt.plot function is used, the variable values are 'total_bill' and 'tip' are pulled directly from the dataframe 'df' using the loc function
#The values are plotted as green triangles and labeled original data
plt.plot(df.loc[:,'total_bill'], df.loc[:,'tip'], 'g^' , label='Original data')
#The next three lines adds a title and labels the x & y axis
plt.title('Tip vs Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
#plt.legend() displayed the legend
plt.legend()
#plt.show() displays the plot
plt.show()
```

The scatter plot show a direct or positive relationship between **total_bill** and **tip**. The higher the total_bill is the higher the tip is likely to be, similarily a low total_bill would result in a lower tip.

### Linear Regression Cost

We will use linear regression analysis to demonstrate there is a positive linear relationship between the independent variable **total_bill** and the dependent variable **tip**. To do this we will calculate and plot a best fit line that is represent the trend of each x & y point of these variables in the tips dataset. This will be represented by a best fit line.

The best line is determined by minimising the following summation formula of all the points of the scatter plot this method is called the **least squares fitting** formula. 

$\sum_i  (y_i - mx_i - c)^2$

Each variable of the formula and its description is given below.

- i = number of x & y values in the dataset that have to be squared and summed above.<br>
- x = total_bill, the value of y (tip) is dependent on the value of x. <br>
- y = tip value<br>
- m = slope of the line<br>
- c = Value of Y axis when x = 0 or Y intercept.<br>

$y_i - mx_i - c$ is determined mathematically from the formula $y_i = mx_i + c$ which is a straight line with a slope of **m, c** is a constant, i.e the value of y when x = 0 or Y intercept.  

$y_i - mx_i - c$ is the measured value from the dataset so for each value of y_i there is a corresponding value x_i.

$y_i = mx_i + c$ is the modeled value this is the straight line

The summation formula is the sum of all the distances each value of x & y (i) or points on the scatter plot is away from a straight line, the (i) or point on the scatter plot can be either above or below the line. 

For each value of i the summation is performed. The best fit line will have the minimum distance from all the points on the plot as a whole. As some points will be above the line and some points will be below the line there will be negative numbers this will not yield the desired result. One possible solution is to take the absolute value from the negative numbers and use it as a positive value, it is easier to simply square all the numbers. If you square a negative or even positive number you will get a positive number. 

We need to find out the distance each dot is from a line and sum this total, the best line will be the lowest value or **Cost(m, c)** of the sum total of each dot's distance from the line. For example if all the dots were on a straight line then the cost would be zero hence the name **least squares fitting** formula.

From the data we need to find the best value for m & c that will yield the lowest cost and therefore the best fit line to the values total_bill and tip.


### Minimising the Cost Calculation using Calculus

The following calculation can be carried to find the miniminm cost for m. When you find m you can calculate c.

```python
#To find the best values of m for the variables total_bill and tip the following calculation is performed

#Assign total_bill and tip values to arrays
total_bill_values = df.loc[: , "total_bill"]
tip_values = df.loc[: , "tip"]

#Find the mean value of both total_bill & tip and assign this value to two variables
avg_totalbill = np.mean(total_bill_values)
avg_tip = np.mean(tip_values)

#Subtract the average or mean values from the original 'total_bill' and 'tip' values for zero values and assign these values to be the zero variables values.
total_bill_values_zero = total_bill_values - avg_totalbill
tip_values_zero = tip_values - avg_tip

#The best m & c is found from the following formula
m = np.sum(total_bill_values_zero * tip_values_zero) / np.sum(total_bill_values_zero * total_bill_values_zero)
# Use m from above to calculate the best c.
c = avg_tip - m * avg_totalbill

#Prints the calculated variable values m & c using formatting of at least 5 digits wide and 8 digits to the right of the decmial point
print("The best value of m is %5.8f and c is %5.8f." % (m, c))
```

### Minimising the Cost Calculation using Python Numpy polyfit function

The numpy polyfit function calculates the m and c values.


```python
#Numpy function that calculates the best value for m & c for a given set of variables. The 1 gives a polynomial of degree 1
np.polyfit(total_bill_values, tip_values, 1)
```

As you can see the **np.polyfit** function calculates the same value for m & c as the Calculus calculation above. 

### Calculating the Cost
To calculate the cost of the values of **total_bill** and **tip** the following calculation can be performed from the resulting m & c values
The Cost can be calculated using the following formula

Cost(m,c) = $\sum_i  (y_i - mx_i - c)^2$

This is given by the code below



```python
cost = np.sum([(total_bill_values[i] - m * tip_values[i] - c)**2 for i in range(total_bill_values.size)])
print("The cost is %8.2f" % (cost))
```

```python
#Plot the best fit line on the scatter plot used previously plotted original data variables total_bill and tip
#This is a magic command to ensure the plot is displayed in Jupyter notebook
%matplotlib inline
#The plt.plot function is used, the variable values are 'total_bill' and 'tip' are pulled directly from the assigned variables in the calulation above 
#The values are plotted as green triangles and labeled original data
plt.plot(total_bill_values, tip_values , 'g^' , label='Original data')
#The best fit line is plotted from the formula 'y = mx + c' using the calculated m & c values for each x, y values, this is plotted as a red line and labeled as Best fit line. 
plt.plot(total_bill_values, m * total_bill_values + c, 'r-' , label='Best fit line')
plt.title('Tip vs Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
#plt.legend() displayed the legend
plt.legend()
#plt.show() displays the plot
plt.show()
```
